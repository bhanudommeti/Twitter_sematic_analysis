```python
import pandas as pd
import numpy as np

from nltk.tokenize import word_tokenize
from nltk.util import ngrams
```

```python
df1 = pd.read_csv('cleaned_data_very.csv')
df1=df1.dropna()
df1
```

```python
df1_tweet = df1['tweets'].apply(word_tokenize)
df1_tweet
```

```python
df1 = pd.DataFrame({'class':df1['class'], 'tweets': df1_tweet})
df1
```

```python
# Function to generate n-grams
def generate_ngrams(tokens, n):
    return [' '.join(gram) for gram in list(ngrams(tokens, n))]

# Generate unigrams, bigrams, and trigrams for each row
df1['unigrams'] = df1['tweets'].apply(lambda tweet: generate_ngrams(tweet, 1))
df1['bigrams'] = df1['tweets'].apply(lambda tweet: generate_ngrams(tweet, 2))
df1['trigrams'] = df1['tweets'].apply(lambda tweet: generate_ngrams(tweet, 3))

df1.head()
```

```python
df1['four_grams'] = df1['tweets'].apply(lambda tweet: generate_ngrams(tweet, 4))
df1
```
